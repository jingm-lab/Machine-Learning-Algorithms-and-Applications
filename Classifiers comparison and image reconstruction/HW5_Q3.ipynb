```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
blue_points = np.array([[0, 0], [2, 2]])
red_points = np.array([[0, 3]])

h = 0.5
red_query = np.array([[h, 1]])

x_vals = np.linspace(-0.2, 6, 100)
y_blue = x_vals

y_red = x_vals + 1 - h

y_decision = x_vals + 0.5 / 2

plt.figure(figsize=(6, 3.5))

plt.scatter(blue_points[:, 0], blue_points[:, 1], color="blue", marker="+", s=100)
plt.scatter(
    red_points[:, 0],
    red_points[:, 1],
    edgecolor="red",
    facecolors="none",
    marker="o",
    s=100,
)
plt.scatter(
    red_query[:, 0],
    red_query[:, 1],
    edgecolor="red",
    facecolors="none",
    marker="o",
    s=100,
)

plt.plot(x_vals, y_blue, "b-", label="Margin Boundary")
plt.plot(x_vals, y_red, "r--", label="Margin Boundary")
plt.plot(x_vals, y_decision, "g-", label="Decision Boundary")

plt.xlim(-0.2, 6)
plt.ylim(-0.2, 3.5)
plt.legend()
_ = plt.xticks(np.arange(0, 6, 1))
```
